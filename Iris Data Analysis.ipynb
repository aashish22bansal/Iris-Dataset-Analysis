```python
import pandas as pd
```

```python
from sklearn.datasets import load_iris
```

```python
data = load_iris()
```

```python
data
```

```python
# Access the features and target variable
X = data.data  # Features (sepal length, sepal width, petal length, petal width)
y = data.target  # Target variable (species: 0 for setosa, 1 for versicolor, 2 for virginica)
```

```python
# Print the feature names and target names
print("Feature names:", data.feature_names)
print("Target names:", data.target_names)
```

```python
sensitive_features = data.feature_names
```

```python
automl = AutoML(
    algorithms=["Xgboost"],
    train_ensemble=True,
    start_random_models=3,
    hill_climbing_steps=3,
    top_models_to_improve=2,
    fairness_threshold=0.8,
    explain_level=1
)
```

```python
automl.fit(X, y)
```
